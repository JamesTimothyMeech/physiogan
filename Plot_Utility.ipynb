```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
titles = ['1X', '10X', '30X', '50X']
values = [71.43, 81.9, 86.67, 89.52]
baseline = 98.1
```

```python
np.arange(len(titles))
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.gca()
ax.set_ylim(0, 100)
#plt.axhline(y=baseline, linestyle='--', color='green')
plt.plot(np.arange(len(titles))+1, 100*np.array(values)/baseline)
ax.set_xticks(np.arange(len(titles))+1)
ax.set_xticklabels(titles)
ax.set_xlabel('Size of Synthetic Training Set', fontsize=28)
ax.set_ylabel('Utility', fontsize=28)
fig.savefig('ecg_utility_plot.png')
```
